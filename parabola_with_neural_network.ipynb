```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import psycopg2
import os
%matplotlib inline
plt.style.use('ggplot')
pd.set_option('display.max_columns', None)
```

```python
def sigma(x,a=1,b=0):
    y = x*a+b
    if y < 0: 
        return 0
    return y
    
```

```python
x = np.linspace(-10,10,100)
y = [sigma(xi) for xi in x]
plt.plot(x,y)
```

```python
f, axs = plt.subplots(3,1,figsize=(5,5))

a = 5
b = -10
x = np.linspace(-10,10,100)
y = [sigma(xi,a,b) for xi in x]

axs[0].plot(x,y,color='blue',label=r'$a=-10\ b=5$')
axs[0].set_ylim(-2,40)
axs[0].set_xlim(0,10)
axs[0].set_xlabel('')
axs[0].set_ylabel(r'$\sigma(a+bx)$')
axs[0].legend(loc='upper left')

y = [sigma(xi,a,2*b) for xi in x]
axs[1].plot(x,y,color='blue',label=r'$a=-10\ b=5$')
axs[1].set_ylim(-2,40)
axs[1].set_xlim(0,10)
axs[1].set_xlabel('')
axs[1].set_ylabel(r'$\sigma(2a+bx)$')
axs[1].legend(loc='upper left')

y = [sigma(xi,a,3*b) for xi in x]
axs[2].plot(x,y,color='blue',label=r'$a=-10\ b=5$')
axs[2].set_ylim(-2,40)
axs[2].set_xlim(0,10)
axs[2].set_xlabel('x')
axs[2].set_ylabel(r'$\sigma(3a+bx)$')
axs[2].legend(loc='upper left')

plt.subplots_adjust(hspace = 0.4)

plt.savefig('ReLUs.pdf')
```

```python
f, axs = plt.subplots(1,1,figsize=(5,5))
a = 5
b = -10
x = np.linspace(0.1,10,100)
y = [sigma(xi,a,b)+sigma(xi,a,2*b)+sigma(xi,a,3*b) for xi in x]
plt.plot(x,y,color='blue',label='NN')
plt.plot(x,x**2,color='red',label=r'$x^2$')
plt.legend(loc='upper left')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('quadratic.pdf')
```
